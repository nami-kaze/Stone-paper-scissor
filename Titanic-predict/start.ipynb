```python
import pandas as pd
data = pd.read_csv('titanic.csv')
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mode()[0], inplace = True)
data['Cabin'].fillna(data['Cabin'].mode()[0], inplace = True)
data['Fare'].fillna(data['Fare'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
#drop column:
data.drop('PassengerId', inplace=True, axis='columns')
data.drop('Name', inplace=True, axis='columns')
```

```python
data.isnull().sum()
```

```python
data['Sex'] = data['Sex'].map({'male':0,'female':1}).astype(int)
```

```python
new_train = data.iloc[:614]
new_test = data.iloc[614:]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(2, 4, figsize=(16, 10))

sns.countplot(x='Survived', data=new_train, ax=ax[0][0])
sns.countplot(x='Pclass', data=new_train, ax=ax[0][1])
sns.countplot(x='Sex', data=new_train, ax=ax[0][2])
sns.countplot(x='Age', data=new_train, ax=ax[0][3])
sns.countplot(x='SibSp', data=new_train, ax=ax[1][0])
sns.countplot(x='Parch', data=new_train, ax=ax[1][1])
sns.countplot(x='Ticket', data=new_train, ax=ax[1][2])
sns.countplot(x='Fare', data=new_train, ax=ax[1][3])

plt.tight_layout()
plt.show()
```
